import pandas as pd
import numpy as np
from scipy import stats

# Função para calcular o limite inferior e superior do Intervalo de Confiança (90%)
def intervalo_confianca_90(dados_coluna):
    n = len(dados_coluna)
    if n < 2: return np.nan  # Precisa de pelo menos 2 dados para calcular
    
    media = np.mean(dados_coluna)
    erro_padrao = stats.sem(dados_coluna) # Erro padrão da média
    
    # Para 90% de confiança, usamos o nível de significância de 0.10
    # O cálculo usa a distribuição t de Student
    intervalo = stats.t.interval(0.90, n-1, loc=media, scale=erro_padrao)
    return intervalo

```python
def intervalo_confianca_90(dados_coluna):
    n = len(dados_coluna)
    if n < 2: return np.nan

    media = np.mean(dados_coluna)
    erro_padrao = stats.sem(dados_coluna)

    # Cálculo do intervalo
    intervalo = stats.t.interval(0.90, n-1, loc=media, scale=erro_padrao)
    return intervalo
```

```python
#DEIXAR A FUNÇÃO PRONTA PARA APLICAR ELA NA PRÓXIMA FASE
def preencher_nivel(gestor,nivel):
  if gestor==1:
    return "pessoa gestora"
  else:
    return nivel
```

```python
#1. Lambda
#É uma função rápida. Em vez de criar um manual de instruções inteiro (def), você escreve uma ordem simples em uma única linha.
#Exemplo: lambda x: x + 1 (Apenas diz: "pegue algo e some um").
#2. Apply
#É o trabalhador. Ele pega essa sua ordem (o lambda) e sai passando ela em cada item de uma coluna ou lista, um por um.
#Na prática: #Se você tem uma coluna de nomes e quer deixar todos em maiúsculo: O apply percorre a lista de nomes.
#O lambda transforma cada nome que o apply entrega para ele.
#Em resumo: O Apply é o motor que leva a função, e o Lambda é a regra simples que diz o que fazer.

#aqui vamos criar a coluna que vai receber o valor da função

dados['NOVO NIVEL']=dados.apply(lambda x: preencher_nivel(x['GESTOR?'], x ['NIVEL']), axis=1)
```

```python
dados['NOVO NIVEL'].value_counts()
```

Imagine que você tem uma caixa cheia de fichas (seu banco de dados). Cada ficha pertence a um funcionário e tem várias informações escritas nela, como o nome do chefe e o cargo.

O que essa linha de código está fazendo é contratar um ajudante robô para organizar essas fichas. Funciona assim:

dados['NOVO NIVEL'] = ...: Você está criando um espaço novo na ficha chamado "NOVO NIVEL" para o robô escrever o resultado.

dados.apply(..., axis=1): Você dá uma ordem para o robô: "Pegue esta pilha de fichas e olhe uma por uma, da primeira até a última". (O axis=1 avisa que ele deve ler a ficha na horizontal, linha por linha).

lambda x:: É o jeito de dizer: "Ei robô, para cada ficha (x) que você pegar, faça o seguinte..."

preencher_nivel(x['GESTOR?'], x['NIVEL']): Aqui é a tarefa secreta. O robô olha na ficha o que está escrito em 'GESTOR?' e o que está escrito em 'NIVEL'. Ele leva esses dois dados para uma "máquina de carimbar" (que é a função preencher_nivel) e ela devolve um novo valor para ele escrever no espaço novo.

Resumindo a história:
É como se você desse um balde de lápis de cor para o robô e dissesse:

"Para cada criança da fila, olhe a cor da camiseta dela e a cor do sapato. Misture as duas e pinte o chapéu dela com a cor nova."

O robô faz o trabalho repetitivo de olhar ficha por ficha para você não ter que fazer na mão!

```python
def cramer_coeficiente(coluna1, coluna2):
    # 1. Cria a tabela de frequência
    tabela_cruzada = pd.crosstab(coluna1, coluna2)

    # 2. Pega o valor de Chi2 (o primeiro valor retornado pela função)
    # chi2_contingency retorna (chi2, p, dof, expected)
    chi2 = chi2_contingency(tabela_cruzada)[0]

    # 3. Soma TOTAL de observações (n)
    # Usar .values.sum() garante que somamos a matriz inteira, não só colunas
    n = tabela_cruzada.values.sum()

    # 4. Menor dimensão da tabela (r ou c) menos 1
    # Se a tabela é 5x3, o shape é (5, 3), o min é 3, então mini = 2
    mini = min(tabela_cruzada.shape) - 1

    # 5. Fórmula do V de Cramer
    v_cramer = np.sqrt(chi2 / (n * mini))

    return v_cramer
```

```python
cramer_coeficiente(dados['COR/RACA/ETNIA'],dados['NIVEL DE ENSINO'])
```

📌Que tal um desafio? Faça a correlação entre nível de ensino e gênero😁

```python
import pandas as pd
import numpy as np
from scipy.stats import chi2_contingency
```

```python
def cramer_v(coluna1, coluna2):
    tabela = pd.crosstab(coluna1, coluna2)
    chi2 = chi2_contingency(tabela)[0]
    n = tabela.values.sum()
    mini = min(tabela.shape) - 1
    return np.sqrt(chi2 / (n * mini))
```

```python
# Calculando a correlação
resultado_cramer = cramer_v(dados['GENERO'], dados['NIVEL DE ENSINO'])
```

```python
print(f"O V de Cramer entre Gênero e Nível de Ensino é: {resultado_cramer:.4f}")
```

```python
# Para visualizar a distribuição real (em porcentagem por gênero)
distribuicao = pd.crosstab(dados['GENERO'], dados['NIVEL DE ENSINO'], normalize='index') * 100
print("\nDistribuição Percentual (%) por Gênero:")
display(distribuicao)
```


#Como interpretar o "desafio":
O V de Cramer vai te dar a intensidade dessa relação:

Próximo de 0: Indica que o gênero e o nível de ensino são independentes (não há uma tendência específica de um gênero estudar mais que o outro).

Próximo de 1: Indica uma forte dependência (ex: um gênero concentra-se massivamente em um nível de ensino específico).

O que observar na tabela:
Ao rodar o pd.crosstab com normalize='index', você conseguirá ver, por exemplo:

"Das pessoas que se identificam como mulheres, qual a % que possui Pós-graduação?"

"Essa % é muito diferente da % observada entre os homens?"

Se o seu V de Cramer for baixo (ex: 0.05), significa que a distribuição de ensino é bem parecida entre os gêneros. Se for alto (ex: 0.30+), há uma desigualdade ou tendência clara que merece ser investigada no seu projeto!
